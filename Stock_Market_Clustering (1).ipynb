```python
data['Delta_Close'] = data['Close'].diff()

# Determine if shares are purchased or sold based on the 'Delta'
data['Action'] = ['Shares purchased' if delta > 0 else 'Shares sold' for delta in data['Delta_Close']]
print(data[['Close', 'Volume', 'Delta_Close', 'Action']])
# Count the occurrences of each action
action_counts = data['Action'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(action_counts, labels=action_counts.index, autopct='%1.1f%%', colors=['#66c2a5', '#fc8d62'])
plt.title('Proportion of Shares Purchased vs Shares Sold')
plt.show()

plt.bar(data.Close,data.Delta_Close,color = ['green' if delta > 0 else 'red' for delta in data['Delta_Close']])
plt.xlabel("Close Price")
plt.ylabel("Delta value")
plt.title("Close price v/s Delta")
plt.legend()
```

```python
plt.scatter(data.Volume,data.Delta_Close)
plt.xlabel("Volume")
plt.ylabel("Delta")
plt.title("VOlume v/s Delta")
```

```python
data['Delta_Volume'] = data['Volume'].diff()
data.Delta_Volume

data['Action_Volume'] = ['Volume Increased' if delta > 0 else 'Volume decreased' for delta in data['Delta_Volume']]
print(data[['Close', 'Volume', 'Delta_Close','Delta_Volume' ,'Action','Action_Volume']])
```

```python
# Drop rows with missing values in 'Delta_Volume' or 'Close'
data1 = data.dropna(subset=['Delta_Volume', 'Close'])

# Extract features (X) and target (y)
X = data1[['Delta_Volume']].values  # Feature (independent variable)
y = data1['Close'].values           # Target (dependent variable)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
predicted = model.predict(X)

# Plot the data and the regression line
plt.figure(figsize=(10, 6))
plt.scatter(data1['Delta_Volume'], data1['Close'], color='blue', label='Data Points')
plt.plot(data1['Delta_Volume'], predicted, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Delta Volume')
plt.ylabel('Close Price')
plt.title('Delta Volume vs Close Price with Regression Line')
plt.legend()
plt.show()
```

```python
data = pd.DataFrame({
    'Close': [1963.15, 1964.95, 1963.8, 1960, 1957, 1959, 1961.7, 1961.05, 1959.8, 1960.05, 1962.05, 1964.95, 1964.95, 1962.6, 1962.2, 1960, 1961.6, 1962.55, 1962.7, 1963.5, 1960, 1955, 1954.55, 1954.35, 1952.1, 1952.65, 1956.1, 1956.05, 1956.05, 1953.55, 1953, 1952.15, 1951.35, 1951.9, 1950.8, 1951.6, 1951.9, 1955.35, 1956.55, 1954.5, 1952.9, 1952.45, 1951.75, 1952.95, 1952.35, 1952.25, 1952.2, 1951.1],
    'Volume': [1360, 562, 545, 470, 1832, 1104, 455, 139, 234, 357, 1211, 638, 888, 543, 482, 831, 330, 491, 733, 2732, 452, 3427, 1474, 1119, 1020, 1977, 1265, 453, 1056, 59, 648, 353, 536, 3927, 1020, 67, 4134, 3193, 315, 1514, 358, 92, 87, 94, 254, 66, 91, 121]
})

# Initialize new columns in your DataFrame
data['Volume Change'] = 0  # To track volume change at each step
data['Volume Increased or Decreased'] = ''  # To track whether the volume increased or decreased

# Initialize total volume with the first volume
total_volume = data['Volume'].iat[0]  # Use .iat for accessing the first value
data.iat[0, 2] = total_volume  # Set the initial volume change (index 0, column 2 for 'Volume Change')
data.iat[0, 3] = 'Initial Volume'  # Mark the first row as initial (index 0, column 3 for 'Volume Increased or Decreased')

# Loop through the 'Close' prices and adjust the volume accordingly
for i in range(1, len(data)):
    if data['Close'].iat[i] > data['Close'].iat[i - 1]:  # Price increased
        total_volume += data['Volume'].iat[i]
        data.iat[i, 2] = total_volume  # Update 'Volume Change' (column 2)
        data.iat[i, 3] = 'Increased'  # Update 'Volume Increased or Decreased' (column 3)
    elif data['Close'].iat[i] < data['Close'].iat[i - 1]:  # Price decreased
        total_volume -= data['Volume'].iat[i]
        data.iat[i, 2] = total_volume  # Update 'Volume Change'
        data.iat[i, 3] = 'Decreased'  # Update 'Volume Increased or Decreased'
    else:  # Price remained the same
        total_volume += data['Volume'].iat[i]  # Just add the volume to previous total
        data.iat[i, 2] = total_volume  # Update 'Volume Change'
        data.iat[i, 3] = 'Same Close Price, Volume Added'  # Mark as same price

# Show the updated DataFrame with changes
print(data)
```
